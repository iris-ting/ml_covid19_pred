# Step2: Data Cleaning

### Missing Values ( County-level mean >  Time-based interpolation (per county) > Global median, don't run)

Imputation Strategy Justification:

To ensure contextual accuracy while preserving sample size, we applied a three-step imputation for all non-target predictors:

1. County-level mean imputation: assumes  similar behavior within a county.

2. Time interpolation by county: accounts for smooth week-to-week changes.

3. Global median fallback: ensures complete matrix for modeling.

```python
def contextual_impute(df, target_col, exclude_cols=['time_value', 'geo_value']):
    """
    Impute missing values using 3-step contextual method:
    1. County-level mean
    2. Time-based interpolation (per county)
    3. Global median fallback
    """
    df = df.dropna(subset=[target_col]).copy().reset_index(drop=True)
    df['time_value'] = pd.to_datetime(df['time_value'])
    df = df.sort_values(['geo_value', 'time_value']).reset_index(drop=True)

    features_to_fill = [col for col in df.columns if col not in exclude_cols + [target_col]]

    for col in features_to_fill:
        if df[col].isnull().sum() > 0:
            print(f"[{col}] missing before: {df[col].isnull().sum()}")

            # Step 1: geo mean
            geo_mean = df.groupby('geo_value')[col].transform('mean')
            df[col] = df[col].fillna(geo_mean)
            print(f"[{col}] after Step 1: {df[col].isnull().sum()}")

            # Step 2: interpolate and apply via lookup dictionary
            for geo, group in df.groupby('geo_value'):
                temp_df = group.set_index('time_value')[[col]].copy()
                temp_df[col] = temp_df[col].interpolate(method='time')
                mask = (df['geo_value'] == geo)
                df.loc[mask, col] = temp_df[col].values
            print(f"[{col}] after Step 2: {df[col].isnull().sum()}")

            # Step 3: fallback median
            df[col] = df[col].fillna(df[col].median())
            print(f"[{col}] after Step 3: {df[col].isnull().sum()}")

    return df
```

```python
# Apply the imputation to both datasets
vaccine_data = contextual_impute(df, target_col='smoothed_wcovid_vaccinated')
covid_positive_data = contextual_impute(df, target_col='smoothed_wtested_positive_14d')
```

> Due to sparse and non-continuous time series for many counties, the time-based interpolation step (Step 2) contributed minimally to missing value reduction. Most missing values were recovered using contextual geo-level means and global medians.

### Missing Values ( time-first / geo-first > Global median, run)

Hence, to improve the quality of imputation and respect the nature of different variables, we grouped predictors into three categoriesthose better imputed first temporally, those better imputed spatially, and the restand applied a customized three-step imputation pipeline to each group.

This hybrid approach allowed us to preserve the distinct temporal or geographic patterns present in the data while minimizing bias from global assumptions.

```python
def smart_imputation(df, target_col, exclude_cols=['time_value', 'geo_value']):
    """
    Perform 3-step smart imputation with variable-specific priority:
    - time-first variables: interpolate over time first, then fill by geo mean, then median
    - geo-first variables: fill by geo mean first, then interpolate over time, then median
    - other variables: treated same as geo-first
    """
    df = df.dropna(subset=[target_col]).copy().reset_index(drop=True)
    df['time_value'] = pd.to_datetime(df['time_value'])
    df = df.sort_values(['geo_value', 'time_value']).reset_index(drop=True)

    exclude_from_imputation = exclude_cols + [target_col]

    # Time-priority variables (trend important)
    time_first_vars = [
        col for col in [
            'smoothed_wspent_time_1d', 'smoothed_wpublic_transit_1d',
            'smoothed_wshop_1d', 'smoothed_wwork_outside_home_1d',
            'smoothed_wrestaurant_1d', 'smoothed_wlarge_event_1d',
            'smoothed_wtested_14d', 'smoothed_wcli', 'smoothed_wtested_positive_14d'
        ] if col != target_col
    ]

    # Geo-priority variables (beliefs, attitudes, static context)
    geo_first_vars = [
        col for col in [
            'smoothed_wworried_become_ill', 'smoothed_wvaccine_likely_govt_health',
            'smoothed_wvaccine_likely_friends', 'smoothed_wvaccine_likely_politicians',
            'smoothed_wvaccine_likely_who'
        ] if col != target_col
    ]

    # Other predictors not in above categories
    other_vars = [col for col in df.columns
                 if col not in exclude_from_imputation and col not in time_first_vars and col not in geo_first_vars]

    # Process time-priority variables
    for col in time_first_vars:
        if col in df.columns and df[col].isnull().sum() > 0:
            print(f"[{col}] missing before: {df[col].isnull().sum()}")

            # Step 1: interpolate over time within each geo_value group
            for geo, group in df.groupby('geo_value'):
                temp_df = group.set_index('time_value')[[col]].copy()
                temp_df[col] = temp_df[col].interpolate(method='time')
                mask = (df['geo_value'] == geo)
                df.loc[mask, col] = temp_df[col].values

            print(f"[{col}] after Step 1 (time): {df[col].isnull().sum()}")

            # Step 2: fill with geo-level mean
            geo_mean = df.groupby('geo_value')[col].transform('mean')
            df[col] = df[col].fillna(geo_mean)
            print(f"[{col}] after Step 2 (geo): {df[col].isnull().sum()}")

            # Step 3: fill with global median
            df[col] = df[col].fillna(df[col].median())
            print(f"[{col}] after Step 3 (median): {df[col].isnull().sum()}")

    # Process geo-priority and other variables
    for col in geo_first_vars + other_vars:
        if col in df.columns and df[col].isnull().sum() > 0:
            print(f"[{col}] missing before: {df[col].isnull().sum()}")

            # Step 1: fill with geo-level mean
            geo_mean = df.groupby('geo_value')[col].transform('mean')
            df[col] = df[col].fillna(geo_mean)
            print(f"[{col}] after Step 1 (geo): {df[col].isnull().sum()}")

            # Step 2: interpolate over time within each geo_value group
            for geo, group in df.groupby('geo_value'):
                temp_df = group.set_index('time_value')[[col]].copy()
                temp_df[col] = temp_df[col].interpolate(method='time')
                mask = (df['geo_value'] == geo)
                df.loc[mask, col] = temp_df[col].values

            print(f"[{col}] after Step 2 (time): {df[col].isnull().sum()}")

            # Step 3: fill with global median
            df[col] = df[col].fillna(df[col].median())
            print(f"[{col}] after Step 3 (median): {df[col].isnull().sum()}")

    return df
```

```python
# Apply the imputation to both datasets
vaccine_data = smart_imputation(df, target_col='smoothed_wcovid_vaccinated')
covid_positive_data = smart_imputation(df, target_col='smoothed_wtested_positive_14d')
```

We compared two imputation strategies: a uniform contextual method (geo  time  median) and a smart-imputation strategy that adjusts the fill order based on the nature of each variable. The latter yielded more meaningful fills for temporally patterned features and belief-oriented indicators, which enhanced downstream prediction accuracy.

Our smart imputation strategy allowed us to tailor the imputation method to the structure of each feature. For behavior-based indicators (e.g., visiting restaurants, commuting), temporal interpolation proved effective, filling in over 80% of missing values. In contrast, belief-based variables (e.g., vaccine trust) showed sparse spatial patterns, with county-level means and global medians playing a greater role. This highlights the importance of context-aware imputation in public health survey data.

```python
def detect_outliers(df, excluded_cols=['time_value', 'geo_value']):
    """
    Detect outliers using IQR method
    """
    features = [col for col in df.columns if col not in excluded_cols]
    outlier_summary = {}

    # IQR Method (works well for non-normally distributed data)
    for feature in features:
        Q1 = df[feature].quantile(0.25)
        Q3 = df[feature].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        outliers = df[(df[feature] < lower_bound) | (df[feature] > upper_bound)]
        outlier_count = len(outliers)
        outlier_percent = (outlier_count / len(df)) * 100

        outlier_summary[feature] = {
            'method': 'IQR',
            'count': outlier_count,
            'percentage': outlier_percent,
            'lower_bound': lower_bound,
            'upper_bound': upper_bound
        }

        print(f"{feature}: {outlier_count} outliers ({outlier_percent:.2f}%)")

    # Create outlier summary dataframe
    outlier_df = pd.DataFrame({
        'Feature': list(outlier_summary.keys()),
        'Outlier Count': [outlier_summary[f]['count'] for f in outlier_summary],
        'Outlier Percentage': [outlier_summary[f]['percentage'] for f in outlier_summary],
        'Lower Bound': [outlier_summary[f]['lower_bound'] for f in outlier_summary],
        'Upper Bound': [outlier_summary[f]['upper_bound'] for f in outlier_summary]
    })

    return outlier_df, outlier_summary
```

```python
print("Detecting outliers in vaccine dataset...")
vaccine_outlier_df, vaccine_outlier_summary = detect_outliers(vaccine_data)
print("\nVaccine Dataset Outlier Summary:")
print(vaccine_outlier_df.sort_values('Outlier Percentage', ascending=False).head(10))

print("\nDetecting outliers in COVID positive dataset...")
covid_outlier_df, covid_outlier_summary = detect_outliers(covid_positive_data)
print("\nCOVID Positive Dataset Outlier Summary:")
print(covid_outlier_df.sort_values('Outlier Percentage', ascending=False).head(10))
```


To ensure feature consistency and semantic validity, we clipped all percentage-type indicators to the logical 0100% range.

Although standard outlier detection methods such as the Interquartile Range (IQR) approach were initially considered, we ultimately decided not to treat values beyond statistical bounds as true outliers. This is because the majority of the features in our dataset represent percentages from survey responses (e.g., mask usage, vaccine willingness), which are inherently bounded between 0 and 100 and are self-reported.

Any perceived "outliers" are not necessarily measurement errors, but may reflect valid regional or temporal variation in respondent behavior and belief. As such, we chose to retain these values without applying IQR-based capping or removal, to preserve the integrity of the original responses.

```python
def inspect_outlier_distributions(df, excluded_cols=['time_value', 'geo_value'], top_n=12, figsize=(18, 12)):
    """
    Inspect distribution of numeric features with potential outliers using histograms and boxplots.
    Does not perform any imputation or value clipping.

    Parameters:
    - df: Input dataframe
    - excluded_cols: Columns to exclude (e.g., time or geo identifiers)
    - top_n: Number of top variables (by variance) to visualize
    """
    features = [col for col in df.columns if col not in excluded_cols and pd.api.types.is_numeric_dtype(df[col])]

    # Sort by variance to choose meaningful features to plot
    top_features = df[features].var().sort_values(ascending=False).head(top_n).index.tolist()

    for feature in top_features:
        fig, axes = plt.subplots(1, 2, figsize=figsize)
        sns.histplot(df[feature].dropna(), bins=30, kde=True, ax=axes[0])
        axes[0].set_title(f'Histogram of {feature}')
        sns.boxplot(x=df[feature], ax=axes[1])
        axes[1].set_title(f'Boxplot of {feature}')
        plt.tight_layout()
        plt.show()
```

```python
inspect_outlier_distributions(vaccine_data, excluded_cols=['time_value', 'geo_value'], top_n=12)
inspect_outlier_distributions(covid_positive_data, excluded_cols=['time_value', 'geo_value'], top_n=12)
```

We inspected the distribution of key predictors using histograms and boxplots. While some variables exhibited skewness and apparent "outliers" in boxplots, we interpret these patterns as normal variations rather than true anomalies, especially given the nature of the data (survey-based proportions).

Hence, we decided not to remove or cap outliers. These values carry meaningful signal (e.g., regional or temporal variation).Also use transformations (e.g., log1p, normalization) if needed for modeling, but avoid arbitrary filtering.

```python
def visualize_distributions(df, target_cols, n_cols=3):
    """
    Create histograms and boxplots for features to examine distributions
    """
    features = [col for col in df.columns if col not in ['time_value', 'geo_value']]
    n_features = len(features)
    n_rows = (n_features // n_cols) + (1 if n_features % n_cols > 0 else 0)

    # Histograms
    plt.figure(figsize=(n_cols * 5, n_rows * 4))
    for i, feature in enumerate(features):
        plt.subplot(n_rows, n_cols, i+1)
        plt.hist(df[feature], bins=30, alpha=0.7)
        plt.title(f'Distribution of {feature}')
        plt.tight_layout()
    plt.savefig('feature_distributions.png')

    # Box plots
    plt.figure(figsize=(n_cols * 5, n_rows * 4))
    for i, feature in enumerate(features):
        plt.subplot(n_rows, n_cols, i+1)
        sns.boxplot(y=df[feature])
        plt.title(f'Boxplot of {feature}')
        plt.tight_layout()
    plt.savefig('feature_boxplots.png')

    # Special focus on target variables
    plt.figure(figsize=(12, 5))
    for i, target in enumerate(target_cols):
        plt.subplot(1, len(target_cols), i+1)
        plt.hist(df[target], bins=30, alpha=0.7, color='royalblue')
        plt.title(f'Distribution of {target}')
    plt.tight_layout()
    plt.savefig('target_distributions.png')
```

### Time Trend

```python
def analyze_time_trends(df, target_cols):
    """
    Analyze trends over time for target variables
    """
    # Ensure time_value is datetime
    df['time_value'] = pd.to_datetime(df['time_value'])

    # Get daily averages
    daily_avg = df.groupby('time_value')[target_cols].mean().reset_index()

    # Plot time series
    plt.figure(figsize=(15, 6))
    for i, target in enumerate(target_cols):
        plt.subplot(1, len(target_cols), i+1)
        plt.plot(daily_avg['time_value'], daily_avg[target])
        plt.title(f'Time Trend of {target}')
        plt.xticks(rotation=45)
        plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.savefig('time_trends.png')

    # Check for unusual spikes or drops
    for target in target_cols:
        daily_avg[f'{target}_pct_change'] = daily_avg[target].pct_change() * 100

    # Plot percentage changes to identify anomalies
    plt.figure(figsize=(15, 6))
    for i, target in enumerate(target_cols):
        plt.subplot(1, len(target_cols), i+1)
        plt.plot(daily_avg['time_value'][1:], daily_avg[f'{target}_pct_change'][1:])
        plt.title(f'Daily % Change in {target}')
        plt.xticks(rotation=45)
        plt.grid(True, alpha=0.3)
        plt.axhline(y=0, color='r', linestyle='-', alpha=0.3)
    plt.tight_layout()
    plt.savefig('time_trend_changes.png')

    return daily_avg
```

```python
print("Analyzing time trends for vaccine data...")
vaccine_daily_avg = analyze_time_trends(vaccine_data, vaccine_targets)

print("\nAnalyzing time trends for COVID positive data...")
covid_daily_avg = analyze_time_trends(covid_positive_data, covid_targets)
```

### Geographical Analysis

```python
def analyze_geographic_patterns(df, target_cols):
    """
    Analyze patterns across different geographic regions
    """
    # Calculate county-level averages
    county_avg = df.groupby('geo_value')[target_cols].mean().reset_index()

    # Get summary statistics by county
    county_stats = {}
    for target in target_cols:
        county_stats[target] = {
            'highest': county_avg.loc[county_avg[target].idxmax()],
            'lowest': county_avg.loc[county_avg[target].idxmin()],
            'std': county_avg[target].std(),
            'mean': county_avg[target].mean()
        }

        print(f"\n{target} Geographic Analysis:")
        print(f"Highest: County {county_stats[target]['highest']['geo_value']} ({county_stats[target]['highest'][target]:.2f})")
        print(f"Lowest: County {county_stats[target]['lowest']['geo_value']} ({county_stats[target]['lowest'][target]:.2f})")
        print(f"Standard Deviation: {county_stats[target]['std']:.2f}")
        print(f"Mean: {county_stats[target]['mean']:.2f}")

    return county_avg, county_stats
```

```python
# Analyze geographic patterns
print("\nAnalyzing geographic patterns for vaccine data...")
vaccine_county_avg, vaccine_county_stats = analyze_geographic_patterns(vaccine_data, vaccine_targets)

print("\nAnalyzing geographic patterns for COVID positive data...")
covid_county_avg, covid_county_stats = analyze_geographic_patterns(covid_positive_data, covid_targets)
```

```python
def plot_correlation_heatmap(df, target_cols, exclude_cols=['time_value', 'geo_value'], filename='correlation_heatmap.png'):
    """
    Compute and plot the correlation heatmap for all numeric features (excluding geo/time columns)
    """
    corr_df = df.drop(columns=exclude_cols).corr()
    plt.figure(figsize=(16, 12))
    sns.heatmap(corr_df, cmap='coolwarm', annot=False, fmt=".2f", square=True)
    plt.title('Feature Correlation Heatmap')
    plt.tight_layout()
    plt.savefig(filename)
    plt.close()

    return corr_df[target_cols].sort_values(ascending=False)
```
